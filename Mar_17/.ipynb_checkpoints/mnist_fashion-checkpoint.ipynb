````python
import tensorflow as tf
from tensorflow import keras
print(tf.__version__)
print(keras.__version__)
print(tf.config.list_physical_devices('GPU'))
````

## 以keras載入資料
使用keras.dataests的fashion_mnist資料

訓練模型的資料主要分為三個train, valid, test。
training data 用來調整模型的參數
validation data 用來在訓練中給模型評分，並該改學習速率(這就是為甚麼validation的正確率比testing低)
testing data 用來給最後的模型評分


````python
fashion_mnist=keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test)=fashion_mnist.load_data()

X_valid, X_train=X_train_full[:5000]/255.0, X_train_full[5000:]/255.0
y_valid, y_train=y_train_full[:5000], y_train_full[5000:]
class_names=['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandals', 'Shirt', 'Sneakers', 'Bag', 'Ankle Boot']

X_train.shape
````

## 利用keras建立模型
keras提供多種方法建立模型\
這次只用的方法是先將各層串起來，再用這串定義模型\
另一種方法是先定義模型，在將各層add上去(所謂的Sequential API)\
這兩中方法定義出來的都是一樣的模型，只是我們用的方法更彈性

````python
input_=keras.layers.Input(shape=(28,28), name='input')
flatten=keras.layers.Flatten(name='flatten')(input_)
hidden1=keras.layers.Dense(300, activation='relu', name='hidden1')(flatten)
hidden2=keras.layers.Dense(100, activation='relu', name='hidden2')(hidden1)
output=keras.layers.Dense(10, activation='softmax', name='output')(hidden2)
model=keras.Model(inputs=[input_], outputs=[output], name='fashion_mnist')
model.summary()
````

## 訓練一個tf模型
在丟資料以前，我們先告訴模型我們要如何訓練他\
然後，以fit去訓練他\
epoch指定我們要用這組資料訓練幾次\

````python
model.compile(loss='sparse_categorical_crossentropy',
             optimizer='sgd',
             metrics=['accuracy'])
history=model.fit(X_train, y_train, epochs=10, validation_data=(X_valid, y_valid))
````

## 視覺化結果
良好的呈現資料是十分重要的\
我們常用的主要有兩個library\
pandas和matplotlib\
應用再查就好

````python
import pandas as pd
import matplotlib.pyplot as plt
pd.DataFrame(history.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
````

## 以testing data評估模型

````python
model.evaluate(X_test, y_test)
````

## 使用模型預測資料
我們用以下的code預測前三比的資料

````python
X_new=X_test[:3]
y_proba=model.predict(X_new)
y_proba.round(2)
````
